```python
#import đồ
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import torch
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import linear, relu
```

```python
#load dataset
dataset= pd.read_csv("C://Users//vutuan//Desktop//My-Projects//Machine Learning//Datasets//Medical-Cost-Personal-Datasets.csv")
dataset
```

```python
#chọn feature
X= dataset.iloc[:,[0,1,2,3,6]].values
y= dataset.iloc[:,[4]].values

print(X,y)
```

```python
#encoding data
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        if(X[i][j]=='female'):
            X[i][j]=0.
        if(X[i][j]=='male'):
            X[i][j]=1.
            
for i in range(y.shape[0]):
    for j in range(y.shape[1]):
        if(y[i][j]=='no'):
            y[i][j]=0
        if(y[i][j]=='yes'):
            y[i][j]=1
            
X = X.astype('float64')
y = y.astype('float64')
print(X)
print(y)
```

```python
#chia data thành train và test
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=8)
#chuyển thành mảng tensor
X_train = tf.convert_to_tensor(X_train)
X_test = tf.convert_to_tensor(X_test)
print(X_test,y_test)
```

```python
#cài đạt mạng neuron
model = Sequential(
    [               
        tf.keras.Input(shape=(5,)),
        Dense(16, activation='relu'),
        Dense(6, activation='relu'),
        Dense(1, activation='sigmoid'),

    ], 
)
```

```python
#chọn loss functin và optimizer
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0003),
)
```

```python
#thông tin mạng neuron
model.summary()
```

```python
history = model.fit(
    X_train,y_train,
    epochs=200
)
```

```python
pred=model.predict(X_test)
for i in range(pred.shape[0]):
    for j in range(pred.shape[1]):
        if(pred[i][j]>=0.5):
            pred[i][j]=1
        if(pred[i][j]<=0.5):
            pred[i][j]=0
y_hat=np.hstack((y_test,pred))
err=0
cor=0
for i in range(y_hat.shape[0]):
    if(y_hat[i][0]!=y_hat[i][1]):
        err+=1
    else:
        cor+=1
print('số lỗi:',err)
print('chính xác:',cor)
print(y_hat)
```
